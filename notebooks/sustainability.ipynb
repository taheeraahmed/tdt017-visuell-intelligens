```python
import matplotlib.pyplot as plt
import math as m

def create_kWh_comparison_plot(models, kWh_values, title):
    # Constants for the additional lines
    bmw_kWh = 19.4
    apartment_kWh = m.floor(200/30)

    # Creating the bar plot with additional lines
    plt.figure(figsize=(10, 6))
    plt.bar(models, kWh_values, color='#ff00ff')
    plt.xlabel('Models')
    plt.ylabel('kWh')
    plt.title(f'kWh Consumption Comparison for {title}')

    # Adding the lines
    plt.axhline(y=bmw_kWh, color='#660066', linestyle='--')
    plt.axhline(y=apartment_kWh, color='#660066', linestyle='--')

    # Adding labels for the lines
    plt.text(len(models)-1, bmw_kWh, f'BMW i4 M50 {bmw_kWh} kWh/100km', va='bottom', ha='right', color='#660066')
    plt.text(len(models)-1, apartment_kWh, f'40 kvm apartment {apartment_kWh} kWh/day', va='bottom', ha='right', color='#660066')

    # Adding labels on top of each bar
    for i in range(len(models)):
        plt.text(i, kWh_values[i], f'{kWh_values[i]:.2f}', ha='center', va='bottom')

    # Show the plot
    title = title.replace(" ", "-")
    print(title)
    plt.savefig(f'../images/sus_4_all/{title}.jpg')
```

```python
models = ["Spleen Baseline", "Spleen Affine", "Spleen Noise", "Spleen Gamma", "Spleen UNETR"]
kWh_values = [1.08, 1.43, 1.13, 1.15, 0.93]

create_kWh_comparison_plot(models,kWh_values, "spleen 100 epochs")
```
